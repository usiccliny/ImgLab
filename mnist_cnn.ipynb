```python
from __future__ import print_function
import keras
import os
import numpy as np
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout, BatchNormalization, Conv2D, MaxPooling2D, Flatten
from keras import backend as K
from keras.models import Sequential
from keras.utils import to_categorical
from PIL import Image
import os
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
batch_size = 128
num_classes = 10
epochs = 12

img_rows, img_cols = 28, 28

x_data = []
y_data = []

output_folder = 'digit_images'

for digit in range(num_classes):
    digit_folder = os.path.join(output_folder, str(digit))
    for i in range(1000):
        image_path = os.path.join(digit_folder, f'{digit}_{i}.png')
        if os.path.exists(image_path):
            img = Image.open(image_path).convert('L')
            img = img.resize((img_rows, img_cols))
            img_array = np.array(img)

            img_array = img_array.astype('float32') / 255.0
            
            x_data.append(img_array)

            y_data.append(digit)

x_data = np.array(x_data)
y_data = np.array(y_data)

x_data = x_data.reshape(-1, img_rows, img_cols, 1)
y_data = keras.utils.to_categorical(y_data, num_classes)

x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, random_state=42)

if K.image_data_format() == 'channels_first':
    input_shape = (1, img_rows, img_cols)
else:
    input_shape = (img_rows, img_cols, 1)

model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3),
                 activation='relu',
                 input_shape=input_shape))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation='softmax'))

model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adadelta(),
              metrics=['accuracy'])

model.fit(x_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(x_test, y_test))

score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
